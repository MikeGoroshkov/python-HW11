# Анализ функции средствами sympy

Проанализировать функцию:
f(x) = -5*x**2 + 10*x - 150

Определить корни
Найти интервалы, на которых функция возрастает
Найти интервалы, на которых функция убывает
Построить график
Вычислить вершину
Определить промежутки, на котором f > 0
Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
```

```python
fx = -5*x**2 + 10*x - 150
fx
```

**График функции**

```python
plot(fx)
```

**Корни функции**

```python
solve(fx, x)
```

```python
x1, x2 = solve(fx)
print(round(x1.evalf(), 3))
print(round(x2.evalf(), 3))
```

**Интервалы убывания и возрастания функции**

```python
diff(fx)
```

```python
plot(diff(fx))
```

```python
solve(diff(fx) > 0)
```

```python
solve(diff(fx) < 0)
```

**Промежутки, на котором f > 0 или f < 0**

```python
solve(fx < 0)
```

```python
solve(fx > 0)
```
